Este caso de estudio permite aplicar conceptos clave de la unidad como el

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

df = pd.read_excel('ESPECIE_DATOS.xlsx')
```

## Historigramas

```python
# esta función regresa el histograma de una columna por especie
def crea_histograma(columna):
    species = df['especie'].unique()
    for sp in species:
        plt.hist(df[df['especie'] == sp][columna], bins=10, alpha=0.5, label=sp, edgecolor='black')

    plt.xlabel(columna)
    plt.ylabel('Frecuencia')
    plt.title(f'Histograma de {columna} por especie')
    plt.legend(title='Especie')
    plt.show()
```

### Histograma Pata

```python
crea_histograma('pata')
```

### Histograma abdomen

```python
crea_histograma('abdomen')
```

### Histograma órganos

```python
crea_histograma('organos')
```
